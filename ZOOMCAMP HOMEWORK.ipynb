```python
# Select only columns RAM, Storage, Screen.
# Select only the desired columns

selected_columns = innjoo_laptops[['RAM', 'Storage', 'Screen']]
selected_columns
```

```python
# Get the underlying NumPy array. Let's call it X.
X = selected_columns.values
X
# Now 'X' contains the NumPy array representation of the selected_columns
```

```python
# Compute matrix-matrix multiplication between the transpose of X and X. To get the transpose, use X.T. Let's call the result XTX.

XTX = X.T @ X
XTX

# Now 'XTX' contains the matrix-matrix multiplication.
# This assumes that X is your NumPy array. The X.T @ X expression calculates the matrix-matrix multiplication between the transpose of X and X, and stores the result in the XTX variable.
```

```python
# Compute the inverse of XTX.
inverse_result = np.linalg.inv(XTX)
inverse_result
# Now 'inverse_result' contains the inverse of the result matrix
```

```python
# Create an array y with values [1100, 1300, 800, 900, 1000, 1100].

y = np.array([1100, 1300, 800, 900, 1000, 1100])

print(y)
# This shows that the y array has been successfully created with the specified values.
```

```python
# Multiply the inverse of XTX with the transpose of X, and then multiply the result by y. Call the result w.
w = inverse_result @ X.T @ y

print(w)

# This assumes that inverse_result, X, and y are defined as in the previous steps. The expression inverse_result @ X.T @ y performs the matrix multiplication in the order specified: first multiply the inverse of result with the transpose of X, and then multiply the result by y. Finally, the result is stored in the w variable.
```

```python
# What's the sum of all the elements of the result?
sum_of_result = np.sum(w)

print("The Sum of Result:", sum_of_result)

# This assumes that result is NumPy array. The np.sum(result) expression calculates the sum of all elements in the result array and stores it in the sum_of_result variable.
```

# EDA# 
Look at the final_price variable. Does it have a long tail?

```python
# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(selected_columns_2['final_price'], bins=50, kde=True)
plt.title('Distribution of Final Price')
plt.xlabel('Final Price')
plt.ylabel('Frequency')
plt.show()
```

# The histogram shows the distribution of the final_price variable. Based on the shape of the histogram, we can observe a long right tail.

# Here are some key observations:

# Skewness: The distribution is skewed to the right, meaning there is a longer tail on the right side. This indicates that there are a few high-priced laptops that are significantly higher than the majority of the data.

# Kurtosis: The distribution also appears to be somewhat leptokurtic, meaning it has heavier tails than a normal distribution. This further emphasizes the presence of outliers or extreme values.

# Long tail interpretation:

# A long tail in this context suggests that there are a relatively small number of laptops with very high prices compared to the majority of laptops. This could be due to factors such as premium brands, high-end configurations, or limited availability.

# Overall, the long tail in the final_price distribution indicates that there are some extreme values that might influence the model's performance. Careful consideration of outlier handling and model selection is necessary to ensure accurate predictions.





```python
# Q-Q plot
import scipy.stats as stats

plt.figure(figsize=(10, 6))
stats.probplot(selected_columns_2 ['final_price'], plot=plt)
plt.title('Q-Q Plot of Final Price')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.show()
```

# The Q-Q plot shows the distribution of the final_price variable compared to a normal distribution. Based on the shape of the plot, we can observe a deviation from normality, especially in the tails.

# Here are some key observations:

# Deviation from normality: The blue dots representing the sample quantiles deviate from the red line representing the theoretical quantiles of a normal distribution. This indicates that the distribution of final_price is not perfectly normal.

# Left tail: The deviation is more pronounced in the left tail, suggesting that there are some lower-priced laptops that are more extreme than what would be expected in a normal distribution.

# Right tail: There is also some deviation in the right tail, but it is less pronounced than in the left tail.

# Interpretation:
# The Q-Q plot confirms the findings from the histogram. The distribution of final_price is skewed to the right, and there are some outliers, especially in the lower price range. This suggests that the data is not perfectly normally distributed, which might have implications for modeling.

# Overall, the Q-Q plot provides further evidence that the final_price variable is not normally distributed.

```python
# Shapiro-Wilk test
stat, p = shapiro(selected_columns_2 ['final_price'])
print('Shapiro-Wilk Statistic:', stat)
print('P-value:', p)
```

# The outcome of Shapiro-Wilk test results indicate that the final_price variable is not normally distributed.

# Here's a breakdown of the output:

# Shapiro-Wilk Statistic: (0.8532061559572949)
# P-value: (6.76218720564138e-41)
# The Shapiro-Wilk test is a statistical test used to assess the normality of a sample. The null hypothesis of the test is that the data comes from a normal distribution.

# Interpretation:

# P-value: The p-value is extremely small (6.76218720564138e-41), which is significantly less than the typical alpha level of 0.05. This means we can reject the null hypothesis.

# Conclusion: Therefore, there is strong evidence to suggest that the final_price variable is not normally distributed.  Transforming the final_price variable to make it more normally distributed may be considered
